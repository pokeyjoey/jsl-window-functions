```python
import sqlite3
conn = sqlite3.connect('imdb.db')
```

```python
movies_df.to_sql('movies', conn,
                 index = False, 
                 if_exists = 'replace')
names_df.to_sql('names', conn,
                index = False,
                if_exists = 'replace')
ratings_df.to_sql('ratings', conn, 
                  index = False,
                  if_exists = 'replace')
title_principals_df.to_sql('movie_roles', conn, 
                           index = False,
                           if_exists = 'replace')
```
